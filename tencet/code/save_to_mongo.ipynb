```python
import pandas as pd
import json
from pymongo import MongoClient
import numpy as np
import matplotlib.pyplot as plt
```

## 保存文件到mognodb

```python
def Mongodb(coll_name='user'):
    user='me'
    password='xsj'
    host='localhost'
    port=2008
    db_name='ten'
    client = MongoClient(host, port)
    client[db_name].authenticate(user, password, db_name)
    db=client[db_name]
    col=db[coll_name]
    return client,db,col
```

```python
def save_file_mongo(file,coll_name,step=int(1e6)): 
    '''
    把csv文件保存的到mongodb
    '''
    client,db,col=Mongodb(coll_name)
    
    with open(r'../data/{}'.format(file)) as f :
        for i,line in enumerate(f):
            if i==0:
                columns=line.strip().split(',')
                print('columns:',columns)
            if i:
                contents=line.strip().split(',')
                contents_dict=dict(zip(columns,contents))
                col.insert_one(contents_dict)
                if i%step==0:
                    print(i)
    client.close()   
    print('all done!')
```

```python
save_file_mongo('adFeature.csv','ad')
```

```python
save_file_mongo('train.csv','train')
```

```python
def find(field,coll_name):
    '''
    从coll_name中读取field 字段
    '''
    client,db,col=Mongodb(coll_name) 
    query=col.find({},{'_id':0,'creativeId':1,'%s'%field:1})
    df=pd.DataFrame(list(query))
    df.loc[df[field]=='',field]=np.nan
    client.close()
    return df
```

```python
df=find('advertiserId','ad')
```
